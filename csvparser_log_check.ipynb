```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('log.csv', header = None)
```

```python
x = df[df.columns[2]].to_list()
y = df[df.columns[3]].to_list()
z = df[df.columns[4]].to_list()
w = df[df.columns[5]].to_list()
rows = df[df.columns[0]].to_list()

arma_5_cols = []
mlpack_5_cols = []
rapid_5_cols = []
rows_5_cols = []
fast_5_cols = []

arma_15_cols = []
mlpack_15_cols = []
rapid_15_cols = []
rows_15_cols = []
fast_15_cols = []

arma_25_cols = []
mlpack_25_cols = []
rapid_25_cols = []
rows_25_cols = []
fast_25_cols = []

for i in range(0, 600, 3):
    rows_5_cols.append(rows[i])
    arma_5_cols.append(x[i])
    mlpack_5_cols.append(y[i])
    rapid_5_cols.append(z[i])
    fast_5_cols.append(w[i])
    
for i in range(1, 600, 3):
    rows_15_cols.append(rows[i])
    arma_15_cols.append(x[i])
    mlpack_15_cols.append(y[i])
    rapid_15_cols.append(z[i])
    fast_15_cols.append(w[i])
    
for i in range(2, 600, 3):
    rows_25_cols.append(rows[i])
    arma_25_cols.append(x[i])
    mlpack_25_cols.append(y[i])
    rapid_25_cols.append(z[i])
    fast_25_cols.append(w[i])
```

```python
plt.figure(figsize=(15,10))
# armadillo
plt.plot(rows_5_cols, arma_5_cols, label = "Armadillo")
# mlpack
plt.plot(rows_5_cols, mlpack_5_cols, label = "Mlpack")
# rapid
plt.plot(rows_5_cols, rapid_5_cols, label = "Rapid")
# fast
plt.plot(rows_5_cols, fast_5_cols, label = "Fast")

plt.xlabel('Number of rows')
plt.ylabel('Time')
plt.title('5 Columns')
plt.legend()
plt.savefig('5_cols.png')
```

```python
plt.figure(figsize=(15,10))
# armadillo
plt.plot(rows_15_cols, arma_15_cols, label = "Armadillo")
# mlpack
plt.plot(rows_15_cols, mlpack_15_cols, label = "Mlpack")
# rapid
plt.plot(rows_15_cols, rapid_15_cols, label = "Rapid")
# fast
plt.plot(rows_15_cols, fast_15_cols, label = "Fast")

plt.xlabel('Number of rows')
plt.ylabel('Time')
plt.title('15 Columns')
plt.legend()
plt.savefig('15_cols.png')
```

```python
plt.figure(figsize=(15,10))
# armadillo
plt.plot(rows_25_cols, arma_25_cols, label = "Armadillo")
# mlpack
plt.plot(rows_25_cols, mlpack_25_cols, label = "Mlpack")
# rapid
plt.plot(rows_25_cols, rapid_25_cols, label = "Rapid")
# fast
plt.plot(rows_25_cols, fast_25_cols, label = "Fast")

plt.xlabel('Number of rows')
plt.ylabel('Time')
plt.title('25 Columns')
plt.legend()
plt.savefig('25_cols.png')
```
