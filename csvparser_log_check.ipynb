```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/home/heisenbug/Dev/C++/mlpack/data-frame/log.csv', header = None)
```

```python
x = df[df.columns[2]].to_list()
y = df[df.columns[3]].to_list()
z = df[df.columns[4]].to_list()

arma_5_cols = []
mlpack_5_cols = []
rapid_5_cols = []

arma_15_cols = []
mlpack_15_cols = []
rapid_15_cols = []

arma_25_cols = []
mlpack_25_cols = []
rapid_25_cols = []

for i in range(0, 600, 3):
    arma_5_cols.append(x[i])
    mlpack_5_cols.append(y[i])
    rapid_5_cols.append(z[i])
    
for i in range(1, 600, 3):
    arma_15_cols.append(x[i])
    mlpack_15_cols.append(y[i])
    rapid_15_cols.append(z[i])
    
for i in range(2, 600, 3):
    arma_25_cols.append(x[i])
    mlpack_25_cols.append(y[i])
    rapid_25_cols.append(z[i])
```

```python
plt.figure(figsize=(15,10))
# armadillo
plt.plot(arma_5_cols, label = "Armadillo")
# mlpack
plt.plot(mlpack_5_cols, label = "Mlpack")
# rapid
plt.plot(rapid_5_cols, label = "Rapid")

plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('5 Columns')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,10))
# armadillo
plt.plot(arma_15_cols, label = "Armadillo")
# mlpack
plt.plot(mlpack_15_cols, label = "Mlpack")
# rapid
plt.plot(rapid_15_cols, label = "Rapid")

plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('15 Columns')
plt.legend()
plt.show()
```
