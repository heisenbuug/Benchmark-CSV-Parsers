```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('logs/logFull.csv', header = None)
```

```python
x = df[df.columns[2]].to_list()
y = df[df.columns[3]].to_list()
z = df[df.columns[4]].to_list()
rows = df[df.columns[0]].to_list()

rows_5_cols = []
arma_5_cols = []
mlpack_custom_5_cols = []
mlpack_new_5_cols = []

rows_15_cols = []
arma_15_cols = []
mlpack_custom_15_cols = []
mlpack_new_15_cols = []

rows_25_cols = []
arma_25_cols = []
mlpack_custom_25_cols = []
mlpack_new_25_cols = []

for i in range(0, 600, 3):
    rows_5_cols.append(rows[i])
    arma_5_cols.append(x[i])
    mlpack_custom_5_cols.append(y[i])
    mlpack_new_5_cols.append(z[i])
    
for i in range(1, 600, 3):
    rows_15_cols.append(rows[i])
    arma_15_cols.append(x[i])
    mlpack_custom_15_cols.append(y[i])
    mlpack_new_15_cols.append(z[i])
    
for i in range(2, 600, 3):
    rows_25_cols.append(rows[i])
    arma_25_cols.append(x[i])
    mlpack_custom_25_cols.append(y[i])
    mlpack_new_25_cols.append(z[i])
```

```python
plt.figure(figsize=(8,5))
# armadillo
# plt.plot(rows_5_cols, arma_5_cols, label = "Armadillo")
# mlpack custom 
plt.plot(rows_5_cols, mlpack_custom_5_cols, label = "Mlpack custom CSV parser")
# mlpack new
plt.plot(rows_5_cols, mlpack_new_5_cols, label = "Mlpack new CSV parser")

plt.xlabel('Number of rows')
plt.ylabel('Time')
plt.title('5 Columns')
plt.legend()
plt.savefig('5_cols.png')
```

```python
plt.figure(figsize=(8,5))
# armadillo
# plt.plot(rows_15_cols, arma_15_cols, label = "Armadillo")
# mlpack
plt.plot(rows_5_cols, mlpack_custom_5_cols, label = "Mlpack custom CSV parser")
# rapid
plt.plot(rows_15_cols, mlpack_new_15_cols, label = "Mlpack new CSV parser")

plt.xlabel('Number of rows')
plt.ylabel('Time')
plt.title('15 Columns')
plt.legend()
plt.savefig('15_cols.png')
```

```python
plt.figure(figsize=(8,5))
# armadillo
# plt.plot(rows_25_cols, arma_25_cols, label = "Armadillo")
# mlpack
plt.plot(rows_5_cols, mlpack_custom_5_cols, label = "Mlpack custom CSV parser")
# rapid
plt.plot(rows_25_cols, mlpack_new_25_cols, label = "Mlpack new CSV parser")

plt.xlabel('Number of rows')
plt.ylabel('Time')
plt.title('25 Columns')
plt.legend()
plt.savefig('25_cols.png')
```
