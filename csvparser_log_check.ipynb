```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/home/heisenbug/Dev/C++/mlpack/data-frame/log.csv', header = None)
```

```python
x = df[df.columns[2]].to_list()
y = df[df.columns[4]].to_list()

x_5_cols = []
y_5_cols = []

x_15_cols = []
y_15_cols = []

x_25_cols = []
y_25_cols = []

for i in range(0, 600, 3):
    x_5_cols.append(x[i])
    y_5_cols.append(y[i])    
    
for i in range(1, 600, 3):
    x_15_cols.append(x[i])
    y_15_cols.append(y[i])    
    
for i in range(2, 600, 3):
    x_25_cols.append(x[i])
    y_25_cols.append(y[i])    
```

```python
plt.figure(figsize=(15,10))
# rapidcsv
plt.plot(y_5_cols, label = "Rapid")
# armadillo
plt.plot(x_5_cols, label = "Armadillo")
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('5 Columns')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,10))
# rapidcsv
plt.plot(y_15_cols, label = "Rapid")
# armadillo
plt.plot(x_15_cols, label = "Armadillo")
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('15 Columns')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,10))
# rapidcsv
plt.plot(y_25_cols, label = "Rapid")
# armadillo
plt.plot(x_25_cols, label = "Armadillo")
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.legend()
plt.show()
```
